```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv('books.csv')
books.head()
```

```python
books.columns
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
books.head()
```

```python
books.rename(columns = {
    "Book-Title": 'title',
    "Book-Author": "author",
    "Year-Of-Publication": "year",
    "Publisher": "publisher",
    "Image-URL-L": "image_url"}, inplace = True)
books.head()
```

```python
users = pd.read_csv('users.csv')
users.head()
```

```python
users = users[['User-ID', 'Location', 'Age']]
```

```python
users.rename(columns = {
    "User-ID": "user_id",
    "Location": "location",
    "Age": "age"}, inplace = True)
users.head(2)
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.rename(columns = {
    "User-ID": "user_id",
    "Book-Rating": "rating"}, inplace = True)
ratings.head(2)
```

```python
ratings = ratings[['User-ID', 'ISBN', 'Book-Rating']]
ratings.head()
```

## Data Analyzing

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
ratings['user_id'].value_counts() #count how much rating from a user
```

```python
ratings['user_id'].unique().shape
```

```python
x = ratings['user_id'].value_counts() > 10
x[x].shape
```

```python
y = x[x].index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)] #show all rating by user per book
ratings.head()
```

```python
ratings_books = ratings.merge(books, on = "ISBN")
ratings_books.head()
```

```python
number_rating = ratings_books.groupby('title')['rating'].count().reset_index()
number_rating.head()
```

```python
number_rating.rename(columns = {"rating": "num_of_rating"}, inplace = True)
number_rating.head()
```

```python
final_rating = ratings_books.merge(number_rating, on = 'title')
final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=30]
final_rating.sample(5)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace=True)
final_rating.shape
```

```python
final_rating
```

## Make Pivot Table

```python
book_pivot_table = final_rating.pivot_table(columns='user_id', index='title', values='rating')
book_pivot_table
```

```python
book_pivot_table.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot_table) #no null data
book_sparse
```

## Make the Model

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
# The algorithm results distance and suggestion. n_neighbors is passing how many suggestion
distance, suggestion = model.kneighbors(book_pivot_table.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot_table.index[suggestion[i]])
```
